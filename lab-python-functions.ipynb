```python
products = ["t-shirt","mug","hat","book","keychain"]
```

```python
#1.
inventory = {}
def initialize_inventory(products):

    for product in products:
        quantity = int(input(f"Enter the quantity for {product}:"))
        
        inventory [product] = quantity #assign the product name as the key and the quantity as the value

    return inventory #return full dictionary
    
    
    
```

```python
print(initialize_inventory(products))
```

```python
#2

customer_orders = set()
def get_customer_orders():
    continuing_order ="yes"
    while continuing_order=="yes":
        product = (input(f"Enter a product name:"))
        customer_orders.add(product)
        continuing_order = input (f"Add another?( yes/no):")
    return customer_orders
```

```python
print(get_customer_orders())
```

```python
#3
def update_inventory (customer_orders, inventory):
    for product in customer_orders:
        if product in inventory: 
            inventory[product] -= 1
    return inventory        
    
```
